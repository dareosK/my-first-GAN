```python
# 2. Make function that shows one img
def view_img(num):
    # store image
    image_tensor = image_folder[num][0]
    # return the shape of an image(tensor)
    # image_folder[7][0].size()
    # torch.size() -> retyrns shape
    # torch.ones() -> returns ones all the 1s
    # torch.flatten() -> flattens to one dimension

    # convert tensor into numpy array (tensors are being processed in the GPU)
    image_numpy = image_tensor.cpu().numpy()
    # squeez images
    squeezed_image = np.squeeze(image_numpy)
    # make an axis for the plot 
    plt.imshow(squeezed_image[0])
    # show the plot on the axis
    plt.show()
view_img(59993)
```

```python
# 4. make discriminator
discriminator = nn.Sequential(
#     one layer
    nn.Conv2d(3, 64, kernel_size=4, stride=2, padding=1, bias=False),
    nn.BatchNorm2d(64),
    nn.LeakyReLU(0.2, inplace=True),
#     second layer
    nn.Conv2d(64, 128, kernel_size=4, stride=2, padding=1, bias=False),
    nn.BatchNorm2d(128),
    nn.LeakyReLU(0.2, inplace=True),
#     third layer
    nn.Conv2d(128, 256, kernel_size=4, stride=2, padding=1, bias=False),
    nn.BatchNorm2d(256),
    nn.LeakyReLU(0.2, inplace=True),
#     fourth layer
    nn.Conv2d(256, 512, kernel_size=4, stride=2, padding=1, bias=False),
    nn.BatchNorm2d(512),
    nn.LeakyReLU(0.2, inplace=True),
    nn.Conv2d(512, 1, kernel_size=4, stride=1, padding=0, bias=False),
    nn.Flatten(),
    nn.Sigmoid()
    )
```

```python
# 5. Make Generator
noise_size = 128
generator = nn.Sequential(
    nn.ConvTranspose2d(noise_size, 512, kernel_size=4, stride=1, padding=0, bias=False),
    nn.BatchNorm2d(512),
    nn.ReLU(True),
    nn.ConvTranspose2d(512, 256, kernel_size=4, stride=2, padding=1, bias=False),
    nn.BatchNorm2d(256),
    nn.ReLU(True),
    nn.ConvTranspose2d(256, 128, kernel_size=4, stride=2, padding=1, bias=False),
    nn.BatchNorm2d(128),
    nn.ReLU(True),
    nn.ConvTranspose2d(128, 64, kernel_size=4, stride=2, padding=1, bias=False),
    nn.BatchNorm2d(64),
    nn.ReLU(True),
    nn.ConvTranspose2d(64, 3, kernel_size=4, stride=2, padding=1, bias=False),
    nn.Tanh()
)
```
