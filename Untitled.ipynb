```python
# to interact with the OS
import os
# machine learning like tensorflow
import torch
# optimizer
import torch.optim as optim
# machine vision
import torchvision
# neural networks
import torch.nn as nn
# micro-framework
import pytorch_lightning as pl
# for math
import numpy as np
# to work with batches
from torch.utils.data import DataLoader
# to make transformation
from torchvision.datasets import ImageFolder
import torchvision.transforms as T
# ploter(maks figures)
import matplotlib.pyplot as plt
# viewing utility(to view your data in a grid)
import torchvision.utils as vutils
# login tool
# import neptune 
# to save images and make grids
from torchvision.utils import save_image, make_grid
# to make morphing effects
from IPython.display import HTML
import random

# Set random seed for reproducibility
seed = 69

print("Random Seed: ", seed)
# assign random seeds
random.seed(seed)
torch.manual_seed(seed)
```

```python
# 1. LOAD DATASET AND ADJUST IT
# a. Locate the images directory
images_path = "inputs"

# define the pixel size of the image (128*128)
image_size = 128
# set the batch size (how many images/loop)
batch_size = 64
stats = (0.5, 0.5, 0.5), (0.5, 0.5, 0.5)

# b. transform the images -same size/centercrop/turn to tensor/and normalize
image_folder = ImageFolder(images_path, transform=T.Compose([
    T.Resize(image_size),
    T.CenterCrop(image_size),
    T.ToTensor(),
    T.Normalize(*stats)
]))

# c. make a dataloader (create objects)
# outputs the images(tensors) in batches, shuffles them. uses 8 cores
data_loader = DataLoader(image_folder, 
                        batch_size=batch_size,
                        shuffle=True,
                        num_workers=8)

# Decide which device we want to run on
device = torch.device("cpu")

# Plot some training images
# manually pull data out of the loader
real_batch = next(iter(data_loader))
# define dimensions(8*8)
plt.figure(figsize=(8, 8))
# turn off the axis of the grid
plt.axis("off")
# give a title
plt.title("Training Images")
# show the grid
plt.imshow(np.transpose(vutils.make_grid(real_batch[0].to(device)[:64], padding=2, normalize=True).cpu(),(1,2,0)))
```

## GAN IMPLEMENTATION

```python
# 2. make discriminator
discriminator = nn.Sequential(
#     one layer
#     convolute
    nn.Conv2d(3, 64, kernel_size=4, stride=2, padding=1, bias=False),
#     normalize
    nn.BatchNorm2d(64),
#     make non-linear
    nn.LeakyReLU(0.2, inplace=True),
#     second layer
    nn.Conv2d(64, 128, kernel_size=4, stride=2, padding=1, bias=False),
    nn.BatchNorm2d(128),
    nn.LeakyReLU(0.2, inplace=True),
#     third layer
    nn.Conv2d(128, 256, kernel_size=4, stride=2, padding=1, bias=False),
    nn.BatchNorm2d(256),
    nn.LeakyReLU(0.2, inplace=True),
#     fourth layer
    nn.Conv2d(256, 512, kernel_size=4, stride=2, padding=1, bias=False),
    nn.BatchNorm2d(512),
    nn.LeakyReLU(0.2, inplace=True),
#     fifth layer
    nn.Conv2d(512, 1, kernel_size=4, stride=1, padding=0, bias=False),
#     flattens the channel to a 1-dimensional tensor (linear operation)
    nn.Flatten(),
    nn.Sigmoid()
    )
# We are taking an image tensor and passing it through five layers. 
# Each convolution expands from 3 channels all the way up to 512 channels. 
# The Final convolution outputs 1 channel
# THEN we pass it through a sigmoid function which then outputs the probability from 0 to 1
```

```python
# 3. Make Generator
noise_size = 128
generator = nn.Sequential(
#     one layer
    nn.ConvTranspose2d(noise_size, 512, kernel_size=4, stride=1, padding=0, bias=False),
    nn.BatchNorm2d(512),
    nn.ReLU(True),
#     two layers
    nn.ConvTranspose2d(512, 256, kernel_size=4, stride=2, padding=1, bias=False),
    nn.BatchNorm2d(256),
    nn.ReLU(True),
#     three layers
    nn.ConvTranspose2d(256, 128, kernel_size=4, stride=2, padding=1, bias=False),
    nn.BatchNorm2d(128),
    nn.ReLU(True),
#     four layers
    nn.ConvTranspose2d(128, 64, kernel_size=4, stride=2, padding=1, bias=False),
    nn.BatchNorm2d(64),
    nn.ReLU(True),
#     five layers
    nn.ConvTranspose2d(64, 3, kernel_size=4, stride=2, padding=1, bias=False),
    nn.Tanh()
)
```
