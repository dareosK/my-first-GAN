## IMPORT LIBRARIES

```python
# to interact with the OS
import os
# machine learning like tensorflow
import torch
# optimizer
import torch.optim as optim
# machine vision
import torchvision
# neural networks
import torch.nn as nn
# micro-framework
import pytorch_lightning as pl
# for math
import numpy as np
# to work with batches
from torch.utils.data import DataLoader
# to make transformation
from torchvision.datasets import ImageFolder
import torchvision.transforms as T
# ploter(maks figures)
import matplotlib.pyplot as plt
# viewing utility(to view your data in a grid)
import torchvision.utils as vutils
# login tool
# import neptune 
# to save images and make grids
from torchvision.utils import save_image, make_grid
# to make morphing effects
from IPython.display import HTML
import random

# Set random seed for reproducibility
seed = 69

print("Random Seed: ", seed)
# assign random seeds
random.seed(seed)
torch.manual_seed(seed)
```

```python
# 1. LOAD DATASET AND ADJUST IT
# a. Locate the images directory
images_path = "inputs"

# define the pixel size of the image (128*128)
image_size = 128
# set the batch size (how many images/loop)
batch_size = 64
stats = (0.5, 0.5, 0.5), (0.5, 0.5, 0.5)

# b. transform the images -same size/centercrop/turn to tensor/and normalize
image_folder = ImageFolder(images_path, transform=T.Compose([
    T.Resize(image_size),
    T.CenterCrop(image_size),
    T.ToTensor(),
    T.Normalize(*stats)
]))

# c. make a dataloader (create objects)
# outputs the images(tensors) in batches, shuffles them. uses 8 cores
data_loader = DataLoader(image_folder, 
                        batch_size=batch_size,
                        shuffle=True,
                        num_workers=8)

# Decide which device we want to run on
device = torch.device("cpu")

# Plot some training images
# manually pull data out of the loader
real_batch = next(iter(data_loader))
# define dimensions(8*8)
plt.figure(figsize=(8, 8))
# turn off the axis of the grid
plt.axis("off")
# give a title
plt.title("Training Images")
# show the grid
plt.imshow(np.transpose(vutils.make_grid(real_batch[0].to(device)[:64], padding=2, normalize=True).cpu(),(1,2,0)))
```

## GAN IMPLEMENTATION

```python
# 2. make discriminator
discriminator = nn.Sequential(
#     one layer
#     convolute
    nn.Conv2d(3, 64, kernel_size=4, stride=2, padding=1, bias=False),
#     normalize
    nn.BatchNorm2d(64),
#     make non-linear
    nn.LeakyReLU(0.2, inplace=True),
#     second layer
    nn.Conv2d(64, 128, kernel_size=4, stride=2, padding=1, bias=False),
    nn.BatchNorm2d(128),
    nn.LeakyReLU(0.2, inplace=True),
#     third layer
    nn.Conv2d(128, 256, kernel_size=4, stride=2, padding=1, bias=False),
    nn.BatchNorm2d(256),
    nn.LeakyReLU(0.2, inplace=True),
#     fourth layer
    nn.Conv2d(256, 512, kernel_size=4, stride=2, padding=1, bias=False),
    nn.BatchNorm2d(512),
    nn.LeakyReLU(0.2, inplace=True),
#     fifth layer
    nn.Conv2d(512, 1, kernel_size=4, stride=1, padding=0, bias=False),
#     flattens the channel to a 1-dimensional tensor (linear operation)
    nn.Flatten(),
    nn.Sigmoid()
    )
# We are taking an image tensor and passing it through five layers. 
# Each convolution expands from 3 channels all the way up to 512 channels. 
# The Final convolution outputs 1 channel
# THEN we pass it through a sigmoid function which then outputs the probability from 0 to 1
```

```python
# Create the generator
netD = discriminator.to(device)

# Print the model
print(netD)
```

```python
# 3. Make Generator
noise_size = 128
generator = nn.Sequential(
#     one layer
    nn.ConvTranspose2d(noise_size, 512, kernel_size=4, stride=1, padding=0, bias=False),
    nn.BatchNorm2d(512),
    nn.ReLU(True),
#     two layers
    nn.ConvTranspose2d(512, 256, kernel_size=4, stride=2, padding=1, bias=False),
    nn.BatchNorm2d(256),
    nn.ReLU(True),
#     three layers
    nn.ConvTranspose2d(256, 128, kernel_size=4, stride=2, padding=1, bias=False),
    nn.BatchNorm2d(128),
    nn.ReLU(True),
#     four layers
    nn.ConvTranspose2d(128, 64, kernel_size=4, stride=2, padding=1, bias=False),
    nn.BatchNorm2d(64),
    nn.ReLU(True),
#     five layers
    nn.ConvTranspose2d(64, 3, kernel_size=4, stride=2, padding=1, bias=False),
    nn.Tanh()
)
```

```python
# Create the generator
netG = generator.to(device)

# Print the model
print(netG)
```

## LOSS functions/optimizers

```python
# Initialize BCELoss function
criterion = nn.BCELoss()

# Size of z latent vector (i.e. size of generator input)
nz = 3

# Learning rate for optimizers
lr = 1e-3

# Beta1 hyperparam for Adam optimizers
beta1 = 0.5

# Create batch of latent vectors that we will use to visualize
#  the progression of the generator
fixed_noise = torch.randn(64, nz, 1, 1, device=device)

# Establish convention for real and fake labels during training
real_label = 1
fake_label = 0

# Setup Adam optimizers for both G and D
# shifts the weights
optimizerD = optim.Adam(netD.parameters(), lr=lr, betas=(beta1, 0.999))
optimizerG = optim.Adam(netG.parameters(), lr=lr, betas=(beta1, 0.999))
```

```python
# Training Loop

# Number of training epochs
num_epochs = 50 # Original is 5 on a dataset of 1 million

# Lists to keep track of progress
img_list = []
G_losses = []
D_losses = []
iters = 0

print("Starting Training Loop...")
# For each epoch
for epoch in range(num_epochs):
    # For each batch in the dataloader
    for i, data in enumerate(data_loader, 0):

        ############################
        # (1) Update D network: maximize log(D(x)) + log(1 - D(G(z)))
        ###########################
        ## Train with all-real batch
        netD.zero_grad()
        # Format batch
        real_cpu = data[0].to(device)
        b_size = real_cpu.size(0)
        label = torch.full((b_size,), real_label, device=device)
        # Forward pass real batch through D
        c.view(-1)
        # Calculate loss on all-real batch
        errD_real = criterion(output, label)
        # Calculate gradients for D in backward pass
        errD_real.backward()
        D_x = output.mean().item()

        ## Train with all-fake batch
        # Generate batch of latent vectors
        noise = torch.randn(b_size, nz, 1, 1, device=device)
        # Generate fake image batch with G
        fake = netG(noise)
        label.fill_(fake_label)
        # Classify all fake batch with D
        output = netD(fake.detach()).view(-1)
        # Calculate D's loss on the all-fake batch
        errD_fake = criterion(output, label)
        # Calculate the gradients for this batch
        errD_fake.backward()
        D_G_z1 = output.mean().item()
        # Add the gradients from the all-real and all-fake batches
        errD = errD_real + errD_fake
        # Update D
        optimizerD.step()

        ############################
        # (2) Update G network: maximize log(D(G(z)))
        ###########################
        netG.zero_grad()
        label.fill_(real_label)  # fake labels are real for generator cost
        # Since we just updated D, perform another forward pass of all-fake batch through D
        output = netD(fake).view(-1)
        # Calculate G's loss based on this output
        errG = criterion(output, label)
        # Calculate gradients for G
        errG.backward()
        D_G_z2 = output.mean().item()
        # Update G
        optimizerG.step()

        # Output training stats
        if i % 1000 == 0:
            print('[%d/%d][%d/%d]\tLoss_D: %.4f\tLoss_G: %.4f\tD(x): %.4f\tD(G(z)): %.4f / %.4f'
                  % (epoch, num_epochs, i, len(data_loader),
                     errD.item(), errG.item(), D_x, D_G_z1, D_G_z2))

        # Save Losses for plotting later
        G_losses.append(errG.item())
        D_losses.append(errD.item())

        # Check how the generator is doing by saving G's output on fixed_noise
        if (iters % 250 == 0) or ((epoch == num_epochs-1) and (i == len(dataloader)-1)):
            with torch.no_grad():
                fake = netG(fixed_noise).detach().cpu()
            img_list.append(vutils.make_grid(fake, padding=2, normalize=True))

        iters += 1
```
